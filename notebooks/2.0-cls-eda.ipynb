```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
train_values = pd.read_csv('../data/raw/training_values.csv')
train_labels = pd.read_csv('../data/raw/training_labels.csv')
train_df = train_values.merge(train_labels, on='id')
train_functional = train_df[train_df['status_group']=='functional']
train_nonfunctional = train_df[train_df['status_group']=='non functional']
train_repair = train_df[train_df['status_group']=='functional needs repair']

test_values = pd.read_csv('../data/raw/test_values.csv')
```

Make feature histograms plotting the functional wells and non-functional wells on the same plot to see if any clear separation appears.

```python
for col in train_df.columns:
    if train_df[col].dtypes in ['int64', 'float64']:
#         bin_size = (train_df[col].max() - train_df[col].min())/20
#         manual_bins = range(int(train_df[col].min()), int(train_df[col].max()+bin_size), int(bin_size))
        
        fig = plt.figure(figsize=(8,8))
        ax = fig.add_subplot(1,1,1)
        ax.hist(train_functional[col], color='g', bins=30, alpha=.6)
        ax.hist(train_nonfunctional[col], color='r', bins=30, alpha=.6)
        ax.hist(train_repair[col], color='k', bins=30, alpha=.6)
        plt.suptitle(col)
        plt.show()
```
